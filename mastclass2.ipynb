```python
temp_data['order_delivered']=0
temp_data['order_expected']=0
for i in range(df.shape[0]):
    temp_data['order_delivered'][i]=temp_data['order'][i]['order_delivered_customer_date']
    temp_data['order_expected'][i]=temp_data['order'][i]['order_estimated_delivery_date']
```

```python
for i in range(temp_data.shape[0]):
    
    datetime_obj = datetime.strptime(temp_data['order_delivered'][i], '%Y-%m-%d %H:%M')

# Format the datetime object as a date string
    temp_data['order_delivered'][i] = datetime_obj.strftime('%Y-%m-%d')
```

```python
temp_data
```

```python
temp_data=temp_data.reset_index(drop=True)
```

```python
for i in range(temp_data.shape[0]):
    
    datetime_obj = datetime.strptime(temp_data['order_delivered'][i], '%Y-%m-%d %H:%M')

# Format the datetime object as a date string
    temp_data['order_delivered'][i] = datetime_obj.strftime('%Y-%m-%d')
```

```python
for i in range(temp_data.shape[0]):
    # Parse the date string using the current format
    if len(temp_data['order_delivered'][i]) > 10:
        date = datetime.strptime(temp_data['order_delivered'][i], '%Y-%m-%d %H:%M')
    else:
        date = datetime.strptime(temp_data['order_delivered'][i], '%Y-%m-%d')
    
    # Format the date as 'YYYY-MM-DD'
    formatted_date = date.strftime('%Y-%m-%d')
    
    # Append the formatted date to the list
    temp_data['order_delivered'][i]=formatted_date
```
